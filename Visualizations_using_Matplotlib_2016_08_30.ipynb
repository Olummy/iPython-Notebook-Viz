# Visualizations using Matplotlib
##### ggplot style

```python
%matplotlib inline
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
ts = ts.cumsum()
ts.plot(figsize=(13,7))
```

```python
ts.head(10)
```

```python
type(ts)
```

```python
ts.index
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
df = df.cumsum()
df.plot(figsize=(15,8))
```

```python
df.plot(subplots=True,figsize=(15,8))
```

```python
df.plot(kind='kde',subplots=True,figsize=(15,8))
```

```python
df.plot(kind='hist',subplots=True,figsize=(15,8),bins=20,sharex=False) #Each of the plot has different x-axis
```

```python
df.plot(kind='hist',subplots=True,figsize=(15,8),bins=20) #The plots share the same x-axis
```

```python
df.plot(kind='box',subplots=True,figsize=(15,8)) ### box plot
```

```python
df.plot(kind='bar',figsize=(15,8),subplots=True) ### bar chart
### dataframe df is not suitable for a bar chart
```

### Visualizing data with python is sweet!

```python
### Bar chart
df2 = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df2.plot(kind='bar',figsize=(15,7))
```

```python
df1 = pd.DataFrame(np.random.rand(10,4),columns=['a','b','c','d'])
```

```python
df1.plot.area(subplots=True,figsize=(13,7));
```

```python
df1.plot.area(subplots=True,figsize=(13,7),sharex=False);
```

```python
df4 = pd.DataFrame(np.random.rand(4,2),index=['a','b','c','d'],columns=['X-Series','Y-Series'])
df4.plot.pie(subplots=True,figsize=(13,7))
```

```python
df4 = pd.DataFrame(np.random.rand(4,2),index=['a','b','c','d'],columns=['X-Series','Y-Series'])
df4.plot.pie(subplots=True,figsize=(13,7),autopct='%.2f',fontsize=12)
```

```python
### Bar chart
df2 = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df2.plot(kind='barh',figsize=(8,10))
```

```python
### Stacked Bar chart
df2 = pd.DataFrame(np.random.rand(10, 4), columns=['a', 'b', 'c', 'd'])
df2.plot.bar(stacked=True,figsize=(15,8))
```

# Scatter Matrix Plot

```python
from pandas.tools.plotting import scatter_matrix
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), columns=['a', 'b', 'c', 'd'])
scatter_matrix(df,diagonal='kde',figsize=(13,7),grid=True,marker='*+')
```

### Auto-mpg Data
